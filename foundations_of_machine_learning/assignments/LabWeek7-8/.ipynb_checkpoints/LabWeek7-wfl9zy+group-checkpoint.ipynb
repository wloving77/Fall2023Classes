# Step 1: Clean the Data

```python
data = pd.read_csv("data/MushroomDataset/secondary_data.csv", delimiter=";")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
# drop the columns with excessive nulls, > 20000 feels absurd, others might have a pattern.
potential_pattern_cols = []

for column in data.columns:
    if data[column].isna().sum() >= 20000:
        data.drop(columns=column, inplace=True)
    elif data[column].isna().sum() > 0:
        potential_pattern_cols.append(column)
```

```python
data.isna().sum()
```

```python
data.head()
```

# Now to Begin Setting Up One Hot Encoding and Standard Scaling Pipeline for All Features:

```python
categorical_transformer = Pipeline(steps = [
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

numerical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
y = data["class"]
X = data.drop(columns="class")

categorical_features = X.select_dtypes(include = ["object", "category"]).columns
numerical_features = X.select_dtypes(include = ["int", "float"]).columns
```

# **Prevalence:**

- Actually very good at about a 50/50 split, can't complain, this gives us a good split of training data for both the negative and positive sides.

```python
num_p = y=="p"
num_p = num_p.sum()

prevalence = num_p / len(y)
prevalence
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

# Now to Build Model the Model and Train on the Data

```python
preprocessor.fit(X)
X_transformed = preprocessor.transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_transformed,y, train_size=0.8, random_state=42)
```

```python
model = KNeighborsClassifier(n_jobs = -1, n_neighbors=3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_probs = model.predict_proba(X_test)
```

```python
for pos, neg in y_probs:
    if (pos!=1 and pos!=0):
        print(pos)
```

```python
pos_probs = []
neg_probs = []
for neg, pos in y_probs:
    pos_probs.append(pos)
    neg_probs.append(neg)
```

# Notable Metrics:

# **Metrics:**

```python
def display_metrics(y_test, y_pred, pos_probs):
    print(f"Recall: {recall_score(np.array(y_test), y_pred, pos_label='p')}")
    print(f"Precision: {precision_score(np.array(y_test), y_pred, pos_label='p')}")
    print(f"Accuracy: {accuracy_score(np.array(y_test), y_pred)}")
    print(f"ROC/AUC Score: {roc_auc_score(y_test, pos_probs)}")
    print(f"F1 Score: {f1_score(np.array(y_test), y_pred, pos_label='p')}")
    
    
    from sklearn.metrics import roc_curve, auc
    fpr, tpr, thresholds = roc_curve(y_test, pos_probs, pos_label='p')
    auc_score = auc(fpr, tpr)
    
    print(f"AUC Score: {auc_score}")

display_metrics(y_test, y_pred, pos_probs)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot()
```

# **Part 4:**
- There does not appear to be an obvious correlation with missclassifications. These values are likely tightly packed and have many values near to them, causing a tight race between the distance to neighbors of both classes.
# **Part 5:**
- Adjusting the Threshold:
- Threshold has relatively little impact as a result of the distance computations, the predictions were shown above, they are always either 0/3, 1/3, 2/3, or 3/3 sure you are in the positive class.
  

```python
def adjusted_threshold(threshold, y_test, y_pred, pos_probs):
    adjusted_preds = []
    for i in range(len(pos_probs)):
        if(pos_probs[i]>=threshold):
            adjusted_preds.append(1)
        else:
            adjusted_preds.append(0)

    display_metrics(y_test, y_pred, adjusted_preds)
```

```python
adjusted_threshold(.75, y_test, y_pred, pos_probs)
```

```python
adjusted_threshold(.25, y_test, y_pred, pos_probs)
```
